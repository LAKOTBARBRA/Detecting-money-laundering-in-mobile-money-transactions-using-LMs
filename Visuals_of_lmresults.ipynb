```python
import matplotlib.pyplot as plt
import numpy as np

# Base data
models = ['LogReg', 'RandForest', 'XGBoost', 'NaiveBayes',
          'D-BERT Pooler', 'D-BERT All', 'D-BERT LoRA', 'DistilGPT2 LoRA']
precisions = [75, 87, 91, 88, 96, 98, 99, 82]
recalls = [99, 94, 98, 99, 95, 94, 95, 99]
f1s = [86, 90, 94, 94, 96, 96, 97, 90]
Accuracy = [83, 89, 94, 93, 96, 96, 97, 80]


def plot_radar(values, title, filename, color):
    labels = models
    num_vars = len(labels)
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    angles += angles[:1]
    values += [values[0]]  # Close the radar

    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    ax.plot(angles, values, color=color, linewidth=2)
    ax.fill(angles, values, color=color, alpha=0.25)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, size=10)
    ax.set_yticks([20, 40, 60, 80, 100])
    ax.set_title(title, y=1.08)

    # Annotate points with values
    for i, (angle, value) in enumerate(zip(angles, values)):
        if i < len(labels):  # Skip the closing point
            ax.text(angle, value + 4, f"{value}%", ha='center', va='center', fontsize=9, color='black')

    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.show()

# Plot each metric
plot_radar(precisions.copy(), 'Radar Plot: Model Precision', 'precision_radar.png', 'orange')
plot_radar(recalls.copy(), 'Radar Plot: Model Recall', 'recall_radar.png', 'green')
plot_radar(f1s.copy(), 'Radar Plot: Model F1-score', 'f1_radar.png', 'coral')
plot_radar(Accuracy.copy(), 'Radar Plot: Model Accuracy', 'Accuracy_radar.png', 'purple')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define data
models = [
    'Logistic Regression', 'Random Forest', 'XGBoost', 'Gaussian Naive Bayes',
    'DistilBERT-only pooler', 'DistilBERT-all-layers', 'DistilBERT-LoRA', 'DistilGPT2 LoRA'
]

fpr_values = [24.56, 13.25, 9.56, 11.09, 3.57, 1.36, 0.13, 18.49]

# Create DataFrame
fpr_df = pd.DataFrame({'Model': models, 'FPR': fpr_values})

# Function to assign color based on FPR
def assign_color(fpr):
    if fpr > 20:
        return 'red'
    elif fpr >= 10:
        return 'gold'
    else:
        return 'limegreen'

# Add a color column
fpr_df['Color'] = fpr_df['FPR'].apply(assign_color)

# Sort for better visuals
fpr_df_sorted = fpr_df.sort_values(by='FPR', ascending=True)

# Plot
plt.figure(figsize=(10, 6))
bars = plt.barh(
    fpr_df_sorted['Model'],
    fpr_df_sorted['FPR'],
    color=fpr_df_sorted['Color']
)

# Add labels
for bar, fpr in zip(bars, fpr_df_sorted['FPR']):
    plt.text(fpr + 0.5, bar.get_y() + bar.get_height() / 2,
             f"{fpr:.2f}%", va='center', fontsize=10)

# Titles and labels
plt.title("False Positive Rate (FPR) Across Models", fontsize=14)
plt.xlabel("False Positive Rate (%)")
plt.xlim(0, max(fpr_values) + 5)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.patches import Patch

# Define data
models = [
    'Logistic Regression', 'Random Forest', 'XGBoost', 'Gaussian Naive Bayes',
    'DistilBERT-only pooler', 'DistilBERT-all-layers', 'DistilBERT-LoRA', 'DistilGPT2 LoRA'
]

fpr_values = [24.56, 13.25, 9.56, 11.09, 3.57, 1.36, 0.13, 18.49]

# Create DataFrame
fpr_df = pd.DataFrame({'Model': models, 'FPR': fpr_values})

# Function to assign color based on FPR
def assign_color(fpr):
    if fpr >= 20:
        return 'red'         # High FPR
    elif 10 < fpr < 20:
        return 'gold'        # Moderate FPR
    else:
        return 'limegreen'   # Low FPR

# Add a color column
fpr_df['Color'] = fpr_df['FPR'].apply(assign_color)

# Sort for better visuals
fpr_df_sorted = fpr_df.sort_values(by='FPR', ascending=True)

# Plot
plt.figure(figsize=(12, 6))
bars = plt.barh(
    fpr_df_sorted['Model'],
    fpr_df_sorted['FPR'],
    color=fpr_df_sorted['Color']
)

# Add labels to bars
for bar, fpr in zip(bars, fpr_df_sorted['FPR']):
    plt.text(fpr + 0.5, bar.get_y() + bar.get_height() / 2,
             f"{fpr:.2f}%", va='center', fontsize=10)

# Titles and labels
plt.title("False Positive Rate (FPR) Across Models", fontsize=14)
plt.xlabel("False Positive Rate (%)")
plt.xlim(0, max(fpr_values) + 5)
plt.tight_layout()

# Add legend (key)
legend_elements = [
    Patch(facecolor='red', edgecolor='black', label='High FPR (20%)'),
    Patch(facecolor='gold', edgecolor='black', label='Moderate FPR (10% < FPR < 20%)'),
    Patch(facecolor='limegreen', edgecolor='black', label='Low FPR (<5%)')
]
plt.legend(handles=legend_elements, loc='lower right')

plt.show()
```
